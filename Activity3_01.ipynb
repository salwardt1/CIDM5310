```python
# import pandas and sqlite libraries
import pandas as pd
import sqlite3
```

```python
# list of tables present in supply company database
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'", sqlite3.connect('supply_company.db'))
tables
```

```python
# SQL method to load the table containing the orders into a DataFrame
orders = pd.read_sql("SELECT * FROM Orders",sqlite3.connect("supply_company.db"))
orders
```

```python
# largest customer ID
largest_cust = orders.groupby('Customer_Number').sum()['amount'].sort_values(ascending = False).index[0]
largest_cust
```

```python
# largest customer data
largest_cust_data = pd.read_sql(("SELECT * FROM Customers WHERE Customer_Number = "  + str(largest_cust)),sqlite3.connect("supply_company.db"))
largest_cust_data
```
