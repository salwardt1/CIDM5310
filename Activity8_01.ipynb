```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_url = 'PUF2020final_v1coll.csv'
data_frame = pd.read_csv(file_url)
```

```python
data_frame = data_frame[["REGION","SQFT","BEDROOMS","PRICE"]]
```

```python
data_frame.head(10)
```

```python
# histogram for PRICE
data_frame.PRICE.plot(kind = 'hist');
```

```python
# histogram for SQFT
data_frame.SQFT.plot(kind = 'hist');
```

```python
# scatter plot for PRICE & SQFT
data_frame.plot(kind = 'scatter', y = 'PRICE', x = 'SQFT');
```

```python
# boxplot for BEDROOMS & PRICE
data_frame.boxplot(by = 'BEDROOMS', column = 'PRICE');
```

```python
# scatter chart for BEDROOMS & PRICE
data_frame.plot(kind = 'scatter', x = 'BEDROOMS', y = 'PRICE');
```

```python
# boxplot for REGION & PRICE
data_frame.boxplot(by = 'REGION', column = 'PRICE');
```

```python
# scatter chart for REGION & PRICE
data_frame.plot(kind = 'scatter', x = 'REGION', y = 'PRICE')
plt.show()
```

```python
# horizontal bar chart for REGION & PRICE
pd.pivot_table(data_frame, index = ['REGION']).PRICE.plot(kind = 'barh');
```

```python
# line chart for REGION & PRICE
pd.pivot_table(data_frame, index = ['REGION']).PRICE.plot();
```
